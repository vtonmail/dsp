```python
from __future__ import print_function, division

%matplotlib inline

import numpy as np

import nsfg
import first
import thinkstats2
import thinkplot
```

```python
lst = []
for i in range(1000):
    lst.append(np.random.random())

pmf = thinkstats2.Pmf(lst, label='random floats [0-1]')
cdf = thinkstats2.Cdf(lst, label='random floats [0-1]')
```

```python
thinkplot.Pmf(pmf)
thinkplot.Config(xlabel='random floats [0-1]', ylabel='PMF')
```

```python
thinkplot.Cdf(cdf)
thinkplot.Config(xlabel='random floats [0-1]', ylabel='CDF')
```

The above charts show that: 1) for large ranges, pmf may not show the distribution properly.  Even though we can tell that the distribution is nearly uniform, it is difficult to observe values corresponding each of the random floats. 2) the CDF in the ideal scenario should be a perfect step function that looks linear but we can see a few minor non-uniformities in the step.
